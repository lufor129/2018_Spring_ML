```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
with tf.Session() as sess:
    image_raw_data = tf.gfile.FastGFile("pic/267_Double portrait of Marie and P.S. Krøyer. The couple have portrayed one another.png","rb").read()
    image_data = tf.image.decode_png(image_raw_data)
    image_data = tf.image.convert_image_dtype(image_data,dtype=tf.float32)

    plt.imshow(image_data.eval())
    plt.show()
    
    image_data1 = tf.image.resize_images(image_data,[299,299],method=0)
    plt.imshow(image_data1.eval())
    plt.show()
    
    image_data2 = tf.image.random_brightness(image_data1,max_delta=90. /255.)
    plt.imshow(image_data2.eval())
    plt.show()
```

```python
def process_data(image):
    image_data = tf.image.convert_image_dtype(image,dtype=tf.float32)
    image_data = tf.image.resize_images(image_data,[299,299],method=0)
    image_data = tf.image.random_saturation(image_data,lower=0.5,upper=1.5)
    image_data = tf.image.random_brightness(image_data,max_delta=90. /255.) #亮度
#     image_data = tf.image.random_contrast(image_data,0.5,1.5)
    image_data = tf.image.random_flip_left_right(image_data)
    
    return image_data
```

```python
with tf.Session() as sess:
    image_raw_data = tf.gfile.FastGFile("pic/350_/350_face1.png","rb").read()
    imagedata = tf.image.decode_png(image_raw_data)
    for _ in range(6):
        result = process_data(imagedata)
        plt.imshow(result.eval())
        plt.show()
```
