```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
save_model_path = './logs/1557561207/model/TransferModel'

img = cv2.imdecode(np.fromfile("./TestPic/350_/350_face1.png",dtype=np.uint8),-1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img = cv2.resize(img,(299,299))
img = np.expand_dims(img, axis=0)
print(img.shape)
```

```python
sess = tf.Session()
new_saver = tf.train.import_meta_graph(save_model_path+".meta")
new_saver.restore(sess, save_model_path)
prediction = tf.get_collection('pred_network')[0]

graph = tf.get_default_graph()
x = graph.get_operation_by_name("input_x").outputs[0]
```

```python
label = sess.run(prediction,feed_dict={x:img})
print(label)
```
