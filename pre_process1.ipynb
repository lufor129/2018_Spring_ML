```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
pic_dir = "pic/"
for data in os.walk(pic_dir):
    folders = data[1]
    for folder in folders[:10]:
        full_path = os.path.join(pic_dir,folder)
        img = cv2.imdecode(np.fromfile(full_path+"/"+folder+".png",dtype=np.uint8),-1)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_patterns = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
        faces = face_patterns.detectMultiScale(img,scaleFactor=1.3,minNeighbors=5,minSize=(20, 20))
        count=1
        for (x, y, w, h) in faces:
            newImage = img[y:y+h,x:x+w]
            plt.imshow(newImage)
            plt.show()
            newImagePath = full_path+"/"+folder+"face"+str(count)+".png"
            newImage = cv2.cvtColor(newImage,cv2.COLOR_BGR2RGB)
            #cv2.imwrite(newImagePath,newImage)
            count = count+1
            #cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        #plt.imshow(img)
        #plt.show()
```

```python
#改名
for data in os.walk(pic_dir):
    files_list = data[2]
    for i in files_list:
        name_title = i.split("_")[0]+"_"
        os.rename(pic_dir+i,pic_dir+name_title+".png")
```

```python
pic_dir = "pic/"

for data in os.walk(pic_dir):
    files_list = data[2]
    for i in files_list:
        try:
            full_path = os.path.join(pic_dir,i)
            image_label = i[:-4]
            if not os.path.exists(pic_dir+image_label):
                os.mkdir(pic_dir+image_label)
            shutil.move(full_path,pic_dir+image_label)
            print ('成功將', full_path, '移動至', image_label, '資料夾')
        except:
            print("woring")
```
